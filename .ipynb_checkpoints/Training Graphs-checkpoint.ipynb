```python
dfTrain = pd.read_csv('exp_local/2022.04.29/102153_cic/train.csv')
dfEval = pd.read_csv('exp_local/2022.04.29/102153_cic/eval.csv')

BIN_SIZE = 25

dfTrain['episode_bin'] = dfTrain.episode // BIN_SIZE * BIN_SIZE
dfEval['episode_bin'] = dfEval.episode // BIN_SIZE * BIN_SIZE
```

```python
sns.lineplot(x=dfTrain.episode_bin, y=dfTrain.episode_reward, label='Train')
g = sns.lineplot(x=dfEval.episode_bin, y=dfEval.episode_reward, label='Eval')
```

```python
def plot(exp):
    dfTrain = pd.read_csv(exp+'train.csv')
    dfEval = pd.read_csv(exp+'eval.csv')

    BIN_SIZE = 1 if dfTrain.shape[0] < 500 else 30

    dfTrain['episode_bin'] = dfTrain.episode // BIN_SIZE * BIN_SIZE
    dfEval['episode_bin'] = dfEval.episode // BIN_SIZE * BIN_SIZE
    sns.lineplot(x=dfTrain.episode_bin, y=dfTrain.episode_reward, label='Train')
    g = sns.lineplot(x=dfEval.episode_bin, y=dfEval.episode_reward, label='Eval')
    g.set_xlabel('Episode')
    g.set_ylabel('Reward')
    g.set_title('Expirement '+exp)
    plt.show()
    
# plot(files[0])
```

```python
for f in files:
    plot(f)
```
